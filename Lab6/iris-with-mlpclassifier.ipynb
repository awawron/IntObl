```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv("../input/Iris.csv") #Завантаження даних
data.sample(5) #Показати довільні 5 спостережень
```

```python
data.head(5)
```

Зображення кореляційного поля

```python
data.describe()
```

Нормалізація (z-стандартизація ) даних 


```python
df_norm = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_norm.sample(n=5)
```

```python
df_norm.describe()
```

Зведення строкових категорійних даних до числових міток

```python
target = data[['Species']].replace(['Iris-setosa','Iris-versicolor','Iris-virginica'],[0,1,2])
target.sample(n=5)
```

Формування набору даних в одному DataF

```python
df = pd.concat([df_norm, target], axis=1)
df.sample(n=5)
```

Створюємо дабір даних для навчання та тестування

```python
train, test = train_test_split(df, test_size = 0.3)
trainX = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]# taking the training data features
trainY=train.Species
testX= test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data features
testY =test.Species
trainX.head(5)
```

```python
trainY.head(5)
```

```python
testX.head(5)
```

```python
testY.head(5)
```

Створюємо об'єкт Multi-layer Perceptron для класифікації
[Multi-layer Perceptron](http://http://scikit-learn.org/stable/modules/neural_networks_supervised.html)

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3, 3), random_state=1)
```

```python
clf.fit(trainX, trainY)
```

```python
prediction = clf.predict(testX)
print(prediction)
```

```python
print(testY.values)
```

```python
print('The accuracy of the Multi-layer Perceptron is:',metrics.accuracy_score(prediction,testY))
```
