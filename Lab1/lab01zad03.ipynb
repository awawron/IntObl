# Analysis of Population in Polish Cities
In this notebook, we will analyze the population growth of three major cities in Poland - Gdansk, Poznan, and Szczecin. We will use Python's Pandas library to load and manipulate data, Matplotlib to visualize the data, and NumPy to compute standard deviation and normalization.

## First steps
First of all we import all the neccessary libraries and define some functions that will help us later to, respectively, standardize, normalize and display the data.

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats


def std(arr):
    re = []
    for item in arr:
        re.append((item - stats.mean(arr)) / np.std(arr, ddof=1))
    return re


def norm(arr):
    re = []
    max_value = max(arr)
    min_value = min(arr)
    for item in arr:
        re.append((item - min_value) / (max_value - min_value))
    return re


def plot_cities(y, data1, data2, data3):
    plt.cla()

    plt.title('Ludnosc w miastach Polski')
    plt.xlabel('Lata')
    plt.ylabel('Liczba ludnosci [w tys.]')

    plt.plot(y, data1, color='r', marker='o')
    plt.plot(y, data2, color='g', marker='o')
    plt.plot(y, data3, color='b', marker='o')

    plt.legend(['Gdansk', 'Poznan', 'Szczecin'])

    plt.show()
```

## Data
We have a CSV file with population data of 3 major cities in Poland from the years 1900 to 2000. We will load this file into a Pandas DataFrame for analysis.

In this step we also add some data as per the task. Then we put the data in four separate variables for easier handling.

```python
path = "D:\\Studia\\UG\\Sem4\\IntObl\\Lab1\\miasta.csv"

miasta = pandas.read_csv(path)

data = {
    'Rok': [2010],
    'Gdansk': [460],
    'Poznan': [555],
    'Szczecin': [405]}
df = pandas.DataFrame(data, index=[10])

new_miasta = pandas.concat([miasta, df])

years = new_miasta.Rok
gdansk_data = new_miasta.Gdansk
poznan_data = new_miasta.Poznan
szczecin_data = new_miasta.Szczecin
```

## Visualization
Using the function **plot_cities()** we plot the population of the three cities over time. The function takes in four parameters - the years and population data for Gdansk, Poznan, and Szczecin.

```python
plot_cities(years, gdansk_data, poznan_data, szczecin_data)
```

## Standard Deviation
We will now compute the standard deviation of the population data for the three cities.

```python
std_gdansk_data = std(gdansk_data)
std_poznan_data = std(poznan_data)
std_szczecin_data = std(szczecin_data)

plot_cities(years, std_gdansk_data, std_poznan_data, std_szczecin_data)
```

Standardizing the data gives us a better understanding of the relative growth of each city.

## Normalization
Finally, we will normalize the population data for the three cities.

```python
norm_gdansk_data = norm(gdansk_data)
norm_poznan_data = norm(poznan_data)
norm_szczecin_data = norm(szczecin_data)

plot_cities(years, norm_gdansk_data, norm_poznan_data, norm_szczecin_data)
```

Normalizing the data gives us a better understanding of the growth of each city relative to its own historical data.
